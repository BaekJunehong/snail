```python
import soundfile as sf
from transformers import WhisperProcessor, WhisperForConditionalGeneration

# 프로세서 및 모델 초기화
processor = WhisperProcessor.from_pretrained("openai/whisper-base")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-base")
forced_decoder_ids = processor.get_decoder_prompt_ids(language="korean", task="transcribe")
```

```python
# 음성 파일 경로 리스트
audio_file_paths = [
    "/content/drive/MyDrive/data/whisper 테스트 음원/audio_21_11_17_3596/K00013596-BMG23-L1N2D1-E-K0KK-02164576.wav",
    "/content/drive/MyDrive/data/whisper 테스트 음원/audio_21_11_17_3596/K00013596-BMG23-L1N2D1-E-K0KK-02164794.wav",
    "/content/drive/MyDrive/data/whisper 테스트 음원/audio_21_11_17_3596/K00013596-BMG23-L1N2D1-E-K0KK-02165323.wav"
    # 음성 파일 추가가능
]

# 각 음원 파일에 대한 처리 및 출력
for audio_file_path in audio_file_paths:
    # 음성 파일 읽기
    audio_data, _ = sf.read(audio_file_path, dtype='float32')

    # 입력 특징 생성
    input_features = processor(audio_data.tolist(), sampling_rate=16000, return_tensors="pt").input_features

    # 토큰 IDs 생성
    predicted_ids = model.generate(input_features, forced_decoder_ids=forced_decoder_ids)

    # 텍스트로 디코딩하여 출력
    transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)
    print(f"음성 파일: {audio_file_path}")
    print("텍스트 출력:", transcription)
    print()
```
